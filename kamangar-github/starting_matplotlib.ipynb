# Starting Matplotlib

## What is Matplotlib

* <tt class="docutils literal"><span class="pre">matplotlib</span></tt> is a python 2D plotting library
* <tt class="docutils literal"><span class="pre">matplotlib</span></tt> can be used in Python scripts or in Ipython
* <tt class="docutils literal"><span class="pre">pyplot</span></tt> is module in <tt class="docutils literal"><span class="pre">matplotlib</span></tt> which provides the state-machine interface to the plotting library.

### Simple plots

```python
# Turn on inline plotting so plots will appear in your notebook.
%matplotlib inline  
from matplotlib import pyplot as plt  # Required import
import numpy as np  # import numpy
x = np.linspace(-20,20,500) # Create 100 linearly spaced numbers from -2 to 2
y = x**2 # y is a function of x (y=x**2)
z = 100*np.sin(x) # z is also a function of x (z=x**3)
plt.plot(x,y) # Plot y vs x
plt.plot(x,z) # Plot z vs x on the same plot
plt.show()  # Show the plot
```

### Changing the plot properties

The examples below show how to change properties of the plot

```python
x = np.linspace(-2,2,1000) # Create 100 linearly spaced numbers from -2 to 2
y = x**2 # y is a function of x (y=x**2)
z = x**3 # z is also a function of x (z=x**3)
plt.ion()
plt.figure(figsize=(8,6)) # Create a figure of 4 inches wide by 3 inches tall
plt.plot(x,y, color='red', label="yzxcvbn") # Set the color of the plot to red
plt.plot(x,z, color='green',linewidth=10 , linestyle="--", label="z") # Set line width and style
plt.xlim(-1,1) # set x axis limits
plt.ylim(-2,2) # Set y axis limits
#plt.xticks([-1,0,1])
plt.legend(loc="upper right") # Set the legend 
ax = plt.gca()  # gca stands for 'get current axis'
# Spines are controlling plot boundaries,  and axis positions.
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('top')
ax.spines['bottom'].set_position(('data',1))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.show()  # Show the plot
```

### Multiple Plots (Subplots)

Multiple plots can be created on the same figure by using the <tt class="docutils literal"><span class="pre">subplot</span></tt>

The genral format of the <tt class="docutils literal"><span class="pre">subplot</span></tt> is:

<tt class="docutils literal"><span class="pre">subplot(number_of_rows, number_of_columns, plot_number)</span></tt>

```python
x = np.linspace(-2,2,100) # Create 100 linearly spaced numbers from -2 to 2
y = x**2 # y is a function of x (y=x**2)
z = x**3 # z is also a function of x (z=x**3)
plt.figure(figsize=(8,6)) 
plt.subplot(1,4,1)
plt.plot(x,y, color='red', label="y") # Set the color of the plot to red
plt.subplot(1,4,2)
plt.plot(x,z, linewidth=2 , linestyle="-", label="z1")
plt.legend(loc="upper left") # Set the legend
plt.subplot(1,4,3)
plt.plot(x,z, linewidth=3 , linestyle="--", label="z2", color='green')
plt.legend(loc="upper left") # Set the legend
plt.subplot(1,4,4)
plt.plot(x,z, linewidth=4 , linestyle="-", label="z3")
plt.legend(loc="upper left") # Set the legend
plt.show()
```

# Working with images

To work with images it is best to use <tt class="docutils literal"><span class="pre">Pillow</span></tt> to read and manipulate images and use  <tt class="docutils literal"><span class="pre">matplotlib</span></tt> to display the image

* Note: <tt class="docutils literal"><span class="pre">matplotlib</span></tt> only supports png format. Use <tt class="docutils literal"><span class="pre">Pillow</span></tt> library to read other image formats.





```python
# Import the required modules
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
my_image=Image.open('hunter.jpg') # read the image
my_image= np.asarray(my_image) # Convert the image to numpy array
plt.figure(figsize=(4,6))
image_plot = plt.imshow(my_image) # Show the image
print(my_image.shape) # Print the size of the image
```

```python
# Show each R layer seperately
r_layer = my_image[:,:,0]
plt.figure(figsize=(2,2))
plt.imshow(r_layer,cmap='Reds_r')
```
